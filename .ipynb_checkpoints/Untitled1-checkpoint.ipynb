```python
from nltk import FreqDist

all_ingredients = [ing for sublist in df['ingredients_list'] for ing in sublist]

fdist = FreqDist(all_ingredients)
top_25 = fdist.most_common(25)

ingredients_top = [item[0] for item in top_25]
counts_top = [item[1] for item in top_25]

plt.figure(figsize=(10, 6))
sns.barplot(x=counts_top, y=ingredients_top, hue=ingredients_top, palette='viridis', legend=False)
plt.title("Top 25 Most Common Ingredients")
plt.xlabel("Frequency")
plt.ylabel("Ingredients")
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

ingredient_strs = df['ingredients_list'].apply(lambda x: ' '.join(x))

vectorizer = CountVectorizer()
ingredient_matrix = vectorizer.fit_transform(ingredient_strs)

ingredient_cooccurrence = (ingredient_matrix.T @ ingredient_matrix).toarray()

sns.heatmap(
    ingredient_cooccurrence[:20, :20],
    cmap='viridis',
    xticklabels=vectorizer.get_feature_names_out()[:20],
    yticklabels=vectorizer.get_feature_names_out()[:20]
)

plt.title('Ingredient Co-Occurrence Heatmap')
plt.show()
```

```python
def recommend_recipes(user_ingredients, df, n_neighbors=5, min_matches=2):
    user_ingredients = [i.lower().strip() for i in user_ingredients]

    df = df.copy()

    df['match_count'] = df['ingredients_list'].apply(
        lambda x: sum(any(ui in ing for ing in x) for ui in user_ingredients)
    )

    filtered_recipes = df[df['match_count'] >= min_matches]

    if filtered_recipes.empty:
        return "No recipes found with given ingredients."

    filtered_recipes = filtered_recipes.reset_index(drop=True)

    recipe_ingredients_matrix = filtered_recipes['ingredients_list'].apply(lambda x: set(x))
    user_ingredient_set = set(user_ingredients)

    all_ingredients = sorted(set(
        ing for sublist in recipe_ingredients_matrix for ing in sublist
    ))

    recipe_vectors = lil_matrix((len(filtered_recipes), len(all_ingredients)))
    ingredient_index = {ing: i for i, ing in enumerate(all_ingredients)}

    for i, ingredients in enumerate(recipe_ingredients_matrix):
        for ing in ingredients:
            recipe_vectors[i, ingredient_index[ing]] = 1

    knn = NearestNeighbors(metric='cosine')
    knn.fit(recipe_vectors)

    user_vector = lil_matrix((1, len(all_ingredients)))
    for ing in user_ingredient_set:
        if ing in ingredient_index:
            user_vector[0, ingredient_index[ing]] = 1

    k = min(n_neighbors, len(filtered_recipes))

    distances, indices = knn.kneighbors(user_vector, n_neighbors=k)

    result = filtered_recipes.iloc[indices[0]].copy()

    result = result.drop_duplicates(subset='TranslatedRecipeName')

    result = result.sort_values(
        by=['match_count', 'TotalTimeInMins'], ascending=[False, True]
    )

    result['URL'] = result['URL'].astype(str).str.strip()
    result['YouTube Link'] = result['YouTube Link'].astype(str).str.strip()

    return result[[
        'TranslatedRecipeName',
        'Cuisine',
        'Course',
        'Diet',
        'Servings',
        'TotalTimeInMins',
        'TranslatedIngredients',
        'TranslatedInstructions',
        'YouTube Link',
        'URL'
    ]].head(k)
```

```python
user_ingredients = ['cheese', 'onion', 'chicken', 'ketchup']
recommended_recipes = recommend_recipes(user_ingredients, df, n_neighbors=5)
print(recommended_recipes)
```

```python
from scipy.sparse import save_npz,load_npz
import joblib

def vectorize_ingredients(recipes_combined):
  recipes_combined['ingredients_str'] = recipes_combined['ingredients'].apply(lambda x : ' '.join(x))
  tfidf = TfidfVectorizer(stop_words = 'english')
  recipe_vectors = tfidf.fit_transform(recipes_combined['ingredients_str'])
  return tfidf,recipe_vectors
def train_knn(recipe_vectors,n_neighbors=5):
  knn = NearestNeighbors(n_neighbors=n_neighbors,metric = 'cosine')
  knn.fit(recipe_vectors)
  return knn
```

```python
def save_artifacts(df, tfidf, recipe_vectors, knn):
    joblib.dump(df, "recipes2.pkl")
    joblib.dump(tfidf, "tfidf_vectorizer2.pkl")
    save_npz("recipe_vectors2.npz", recipe_vectors)
    joblib.dump(knn, "knn_model2.pkl")
```

```python
def load_artifacts():
    df = joblib.load("recipes2.pkl")
    tfidf = joblib.load("tfidf_vectorizer2.pkl")
    recipe_vectors = load_npz("recipe_vectors2.npz")
    knn = joblib.load("knn_model2.pkl")
    return df, tfidf, recipe_vectors, knn
```

```python
def recommend_recipes(user_ingredients, tfidf, knn, df, n_neighbors=5):
    user_input = ' '.join([i.lower().strip() for i in user_ingredients])
    
    user_vector = tfidf.transform([user_input])
    
    k = min(n_neighbors, df.shape[0])
    distances, indices = knn.kneighbors(user_vector, n_neighbors=k)
    
    result = df.iloc[indices[0]].copy()
    
    result = result.drop_duplicates(subset='TranslatedRecipeName')
    
    return result[[
        'TranslatedRecipeName',
        'Cuisine',
        'Course',
        'Diet',
        'Servings',
        'TotalTimeInMins',
        'TranslatedIngredients',
        'TranslatedInstructions',
        'YouTube Link',
        'URL'
    ]].head(k)
```

```python
def vectorize_ingredients(df):
    df = df.copy()
    df['ingredients_str'] = df['ingredients_list'].apply(lambda x: ' '.join(x))
    
    tfidf = TfidfVectorizer(stop_words='english')
    recipe_vectors = tfidf.fit_transform(df['ingredients_str'])
    
    return tfidf, recipe_vectors
```

```python
# FIRST run this to CREATE the files
tfidf, recipe_vectors = vectorize_ingredients(df)
knn = train_knn(recipe_vectors)

save_artifacts(df,tfidf,recipe_vectors,knn)

# THEN load
df, tfidf, recipe_vectors, knn = load_artifacts()
```

```python
user_ingredients = ['tomato', 'cheese', 'onion', 'chicken', 'milk']
recommendations = recommend_recipes(user_ingredients, tfidf, knn, df)
print(recommendations)
```
